```python
import requests
import pandas as pd
from io import StringIO

BASE_URL="https://aviationweather.gov/adds/dataserver_current/httpparam"
    
stationString = "KOAK,KHWD,KSFO,KHAF,KSQL,KPAO,KNUQ,KSJC,KRHV,KC83,KTCY"
query = {"dataSource":"metars","requestType":"retrieve","format":"csv","stationString":stationString, "hoursBeforeNow":1}

data = requests.get(BASE_URL, params=query)
```

```python
buffer = StringIO(data.text)
df = pd.read_csv(buffer, skiprows=5)
df = df.drop(["raw_text"], axis=1)
df
```
