```python
import xarray as xr
import rioxarray
import plotly.express as px
import numpy as np
```

```python
xds = xr.open_dataset("../flask/application/static/farsite.nc", decode_coords="all")
xds
```

```python
xds_lonlat = xds.rio.reproject("EPSG:4326")
xds_lonlat
```

```python
xds_lonlat_low = xds_lonlat.coarsen(dim={"x": 16, "y": 16}, boundary="trim").mean()
xds_lonlat_low
```

```python
df = xds_lonlat_low.to_dataframe()
df.reset_index(inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
df.describe()
```

```python
xds_lonlat.to_netcdf("../flask/application/static/farsite_lonlat.nc")
xds_lonlat_low.to_netcdf("../flask/application/static/farsite_lonlat_low.nc")
df.to_csv("../flask/application/static/farsite_lonlat_low.csv", index=False)
```
